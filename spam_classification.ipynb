```python
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
corpus=[]
for i in range(0,len(messages)):
    filter=re.sub('[^a-zA-Z0-9]',' ',messages['message'][i])
    filter=filter.lower()
    filter=filter.split()
    
    filter=[ps.stem(word) for word in filter if word not in set(stopwords.words('english'))]
    filter=' '.join(filter)
    corpus.append(filter)
```

<h1>AVERAGE WORD2Vec</h1>
Avg Word2Vec is an extension of the Word2Vec model that generates vector representations for sentences or documents instead of individual words. It works by taking the average of the vector representations of all the words in a sentence or document to generate a single vector representation for the entire text

```python
import numpy as np
def avg_word2vec(doc):
    #remove out-of-vocalbulary words
    #sent=[word for word in doc if not in model.wv.inex_to_key]
    #print(sent)

    return np.mean([model.wv[word] for word in doc if word  in model.wv.index_to_key],axis=0)
```
